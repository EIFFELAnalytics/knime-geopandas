```python
# Read from input/output table
import pandas as pd
input_table = pd.read_csv('io_table.csv')
```

```python
# CENTROID NODE
from header import df_to_gdf, gdf_to_df

# Read input
gdf = df_to_gdf(input_table)

# Calculate centroids. Note: this overwrites the current geometry column
gdf.geometry = gdf.geometry.centroid

# Output
output_table = gdf_to_df(gdf)
```

```python
# Write to input/output table
output_table.to_csv('io_table.csv', index=False)

# Show please
%matplotlib inline
gdf.plot(figsize=(10,10));
```
